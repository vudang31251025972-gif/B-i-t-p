```python
import folium
import pandas as pd
data = pd.DataFrame({ 'lat': [10.7830141, 10.7770348, 10.7795, 10.7859, 10.7870], 'lon': [106.6948927, 106.695488, 106.6921, 106.6906, 106.6935], 'name': ['UEH A', 'Dinh Độc Lập','Bảo tàng chiến tích chiến tranh','Công viên Lê Văn Thám','Bảo tàng Phụ nữ Nam Bộ']})
m = folium.Map(location=[10.7830141, 106.6948927], zoom_start=16)
for i, row in data.iterrows():
 folium.Marker([row['lat'], row['lon']], popup=row['name']).add_to(m)
m
```
