```python
import pandas as pd
import folium
from folium.plugins import HeatMap
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my_request")
data={'latitude':[10.7562744,10.7565468,10.755264,10.7561278,10.756114,10.7556384],'longitude':[106.6638299,106.665875,106.6629435,106.6616627,106.6627253,106.6646794]}
dia_diem=["cho Ben Thanh, Ho Chi Minh City",
"Dinh doc lap, Ho Chi Minh City",
"Nha tho Duc ba, Ho Chi Minh City",
"Cong vien Tao Dan, Ho Chi Minh City",
"Saigon Centre, Ho Chi Minh City",
"Pho di bo Nguyen Hue, Ho Chi Minh City",
"Benh vien Cho Ray, Ho Chi Minh City",
"Dai hoc y duoc thanh pho Ho Chi Minh, Ho Chi Minh City",
"san bay quoc te Tan Son Nhat, Ho Chi Minh City",
"Landmark 81, Ho Chi Minh City"]
for i in dia_diem:
    location = geolocator.geocode(i)
    if location:
        data['latitude'].append(location.latitude)
        data['longitude'].append(location.longitude)
df = pd.DataFrame(data)
m=folium.Map(location=[10.761327233494937, 106.66828289481502], zoom_start=13)
heat_data = df[['latitude','longitude']].values.tolist()
HeatMap(heat_data).add_to(m)
m
```
